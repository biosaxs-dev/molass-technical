```python
import sys
sys.path.insert(0, '../../../molass-library')
sys.path.insert(0, '../../../molass-legacy')
import os
import matplotlib.pyplot as plt
from molass.Local import get_local_settings
from molass.DataUtils.FolderWalker import walk_folders
from molass.Global.Options import set_molass_options
set_molass_options(flowchange='auto')
local_settings = get_local_settings()
DATA_ROOT_FOLDER = local_settings['DATA_ROOT_FOLDER']
from molass.DataObjects import SecSaxsData as SSD

output_folder = "figs"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for k, path in enumerate(walk_folders(DATA_ROOT_FOLDER)):
    # if k < 2:
    # if k < 28:
    # if k < 42:
    # if k < 52:
    # if k < 75:
    # if k < 76:
    # if k < 89:
    # if k < 90:
    # if k < 95:
    # if k < 97:
    # if k < 125:
    # if k < 145:
    # if k < 155:
    # if k < 161:
    # if k < 166:
    # if k < 187:   # loop?
    # if k < 188:
    # if k < 189:
    # if k < 212:
    # if k < 235:
    if k < 237:
        continue
    print([k], path)
    ssd = SSD(path, remove_bubbles=True)
    if ssd.uv is None:
        print("No UV data found in %s, skipping." % path)
        continue

    trimmed_ssd = ssd.trimmed_copy()
    corrected_ssd = trimmed_ssd.corrected_copy()
    reuslt = corrected_ssd.plot_compact(title=path, baseline=True)
    reuslt.savefig(os.path.join(output_folder, f"{k:03d}_corrected.png"))
    reuslt.close()
    decomposition = corrected_ssd.quick_decomposition(num_components=3)
    result1 = decomposition.plot_components(title=path)
    result1.savefig(os.path.join(output_folder, f"{k:03d}_egh.png"))
    result1.close()
    sdm_decomposition = decomposition.optimize_with_model('SDM')
    result2 = sdm_decomposition.plot_components(title=path)
    result2.savefig(os.path.join(output_folder, f"{k:03d}_sdm.png"))
    result2.close()
    edm_decomposition = decomposition.optimize_with_model('EDM')
    result3 = edm_decomposition.plot_components(title=path)
    result3.savefig(os.path.join(output_folder, f"{k:03d}_edm.png"))
    result3.close()

    if k > 3:
        # break  # Uncomment to stop after processing a few files
        pass
```
