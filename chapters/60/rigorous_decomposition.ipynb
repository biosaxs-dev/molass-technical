# Rigorous Decompositon

```python
from molass import get_version
from molass_data import SAMPLE4
from molass.DataObjects import SecSaxsData as SSD
assert get_version(toml_only=True) >= "0.7.3", "Please update molass to version 0.7.3 or higher."
print("molass version:", )
ssd = SSD(SAMPLE4)
trimmed_ssd = ssd.trimmed_copy()
corrected_ssd = trimmed_ssd.corrected_copy()
corrected_ssd.plot_compact(baseline=True);
```

```python
decomposition = corrected_ssd.quick_decomposition(num_components=3)
rgcurve = corrected_ssd.xr.compute_rgcurve()
decomposition.plot_components(title="Decomposition of Sample4 with Rg Curve", rgcurve=rgcurve)
```

```python
r_decomposition = decomposition.rigorous_decomposition(rgcurve=rgcurve, debug=True)
r_decomposition.plot_components(title="Rigorous Decomposition of Sample4", rgcurve=rgcurve);
```
